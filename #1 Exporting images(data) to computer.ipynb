```python
import sunpy 
import astropy.units as u

from sunpy.net import Fido, attrs as a 
import astropy.time
```

```python
#Fido
#What Fido proivides us interms of instrament Data
```

```python
#a.Instrument 
#Shows us the instraments we are able to use for our Data
```

```python
Sun_Start = "2021-01-17T00:00" #Start Time
```

```python
Sun_End = "2021-01-17T00:01" #End time
```

```python
Sun_Time = a.Time(Sun_Start , Sun_End) 
```

```python
#a.Instrument.aia 
#Specific instrament we use for our Data (AIA)
```

```python
 Data = Fido.search(Sun_Time & a.Instrument.aia & a.Wavelength(94*u.angstrom)) #Gets Data
```

```python
#Data 

#Just stating the variable so that we can see the table below
```

```python
files = Fido.fetch(Data)#, path='data/{instrument}') 
#Converting Data into a File on your computer
```
