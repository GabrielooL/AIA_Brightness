```python
#This is to get our files from computer to Python
List_2 = glob.glob(r'C:\Users\14242\2022 Summer Internship\Sun Pics Part 2 (10 pics)\*')
```

```python
#Helps us with the variables and grates a loop with the givin variable
data_average = [] 
data_median = [] 
for Polar in List_2: 
    Polardata = getdata(Polar) 
    data_average.append(Polardata - np.average(Polardata)) 
    data_median.append(Polardata - np.median(Polardata)) 
```

```python
#Creates an array fro the average
data_average = np.array(data_average)
```
