```python
#The reason why the garph is like this is bc its grapging all the particals on one go 
fig, ax = plt.subplots() 

Fra =  t2['frame'].iloc[iloc_valid]
Sig =  t2['signal'].iloc[iloc_valid]
ax.plot(Fra,Sig) 
plt.ylabel('Signal') 
plt.xlabel('Frame')  


```

```python
data = pd.DataFrame()  

for item in set (t2.particle): 
    sub = t2[t2.particle == item] 
    frame = t2['frame'].iloc[iloc_valid] 
    signal = t2['signal'].iloc[iloc_valid] 
    #particle =t2['particle'].iloc[iloc_valid]  
    for particle,signal,frame in zip(sub.frame[:-1], sub.signal[:-1],sub.particle[:-1],): 
        data = data.append([{'frame': frame, 
                            'signal': signal, 
                            'particle': item 
                            }]) 
data.head()
```

```python
#For data in the DataFrame we can create a nested_dictionary 
#so that we can manipulate data much more effcently.  
mydict = {}  
for light in np.unique(data.particle): 
    d ={}
    #d['particle'] = t.iloc.particle[t.iloc.particle == light] 
    d['frame'] = data.frame[data.particle == light]
    d['signal'] = data.signal[data.particle == light] 
    mydict[light] = d
    
```

```python
for light in mydict.keys(): 
    v_sig = mydict[light]['signal'] 
    v_fra = mydict[light]['frame'] 
    
    plt.plot(v_fra,v_sig)
    
plt.xlabel('frame')
plt.ylabel('signal')
plt.show()
```
