```python
#This Code is to identify a variable for the other 4 files

Files_names= [r"C:\Users\14242\data\AIA\2 Pic Sun.fits", 
                           r"C:\Users\14242\data\AIA\3 Pic Sun.fits", 
                           r"C:\Users\14242\data\AIA\4 Pic Sun.fits", 
                           r"C:\Users\14242\data\AIA\5 Pic Sun.fits"]
```

```python
#This Code forms the bases for makeing a loop for the 4 files. This means that this loop will be used for future codes  

#What this code does gives us is the print of the files names, NOT the Files and the Data it's self
for Sun in Files_names:
    print(Sun)
```

```python
#This code is a combination fo the Loop on top and averages
#Now to acually get the files and the data we need to use sunpy.map/Map 
#Sun is the Variable for the Loop 
#SunMap is the container for the Files and thier Data 
#Cmbining (SunMap) and (Sun) gives is a loop of the files and its data
averages = [] 
for Sun in Files_names: 
    SunMap= sunpy.map.Map(Sun)    
    averages.append(np.average(SunMap.data))
```
